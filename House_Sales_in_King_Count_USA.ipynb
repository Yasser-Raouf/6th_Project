The error you're encountering is due to a string value ('20141013T000000') being present in a column that is expected to contain numeric values. To identify and locate these problematic values, we can use pandas to find non-numeric entries in our DataFrame.

```python
# Define a function to check if a value is convertible to float
def is_convertible_to_float(value):
    try:
        float(value)
        return True
    except ValueError:
        return False

# Apply the function to each column and identify non-convertible values
non_numeric_columns = {}
for col in df.columns:
    if df[col].dtype == 'object':
        non_numeric_columns[col] = df[~df[col].apply(is_convertible_to_float)]

# Display the columns and rows with non-convertible values
for col, rows in non_numeric_columns.items():
    print(f"Column: {col}")
    print(rows)
```

```python
import statsmodels.api as sm

plt.figure(figsize=(8, 6))
sns.regplot(x='sqft_living', y='price', data=df)
plt.title('Relationship between Square Footage of Living and Price')
plt.xlabel('Square Footage of Living')
plt.ylabel('Price')

# Show the plot
plt.show()

correlation = df['sqft_living'].corr(df['price'])
print(f"Correlation coefficient between sqft_living and price: {correlation}")

# Extract the independent variable (sqft_living) and the dependent variable (price)
X = df['sqft_living']
y = df['price']

# Add a constant to the independent variable (to include the intercept in the model)
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Calculate and print the R^2 value
r_squared = model.rsquared
print(f"R^2 value: {r_squared}")

# Print a summary of the regression model (optional, for more details)
#print(model.summary())
```

# Module 5: Model Evaluation and Refinement


Import the necessary modules:


```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
print("done")
```

We will split the data into training and testing sets:


```python
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]    
X = df[features]
Y = df['price']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15)


print("number of test samples:", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```


Create and fit a Ridge regression object using the training data, set the regularization parameter to 0.1, and calculate the R^2 using the test data. Take a screenshot of your code and the value of the R^2. You will need to submit it for the final project.


```python
from sklearn.linear_model import Ridge
```

```python
# Create and fit the Ridge regression model
ridge = Ridge(alpha=0.1)  # Set the regularization parameter (alpha) to 0.1
ridge.fit(x_train, y_train)

# Predict the target variable for the test data
y_pred = ridge.predict(x_test)

# Calculate the R² score using the test data
r_squared = r2_score(y_test, y_pred)
print(f"R^2 score on test data: {r_squared}")
```


Perform a second order polynomial transform on both the training data and testing data. Create and fit a Ridge regression object using the training data, set the regularisation parameter to 0.1, and calculate the R^2 utilising the test data provided. Take a screenshot of your code and the R^2. You will need to submit it for the final project.


```python
# Perform second-order polynomial transformation on the training and testing data
poly = PolynomialFeatures(degree=2)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)

# Create and fit the Ridge regression model using the transformed training data
ridge = Ridge(alpha=0.1)  # Set the regularization parameter to 0.1
ridge.fit(x_train_poly, y_train)

# Predict the target variable for the test data
y_pred = ridge.predict(x_test_poly)

# Calculate the R² score using the test data
r_squared = r2_score(y_test, y_pred)
print(f"R^2 score on test data after polynomial transformation: {r_squared}")
```
